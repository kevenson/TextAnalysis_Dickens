# Text Analysis using TextBlob

```python
from textblob import TextBlob as TB
import codecs
from collections import Counter
```

```python
#file = open('OliverTwist.txt', "r", encoding='iso-8859-15')
def read_file(filename):
    infile = open(filename)# windows users should use codecs.open after importing codecs
    contents = infile.read()
    infile.close()
    return contents
    
```

```python
#print(file.read())
oliver = read_file("OliverTwist_GB.txt")
print(oliver)
```

```python
with open("OliverTwist.txt", "r", encoding='iso-8859-15') as corpus:
    data=corpus.read().replace("\n", "")
#print(data)
OTblob = TB(data)
```

## Sentiment Analysis

```python
OTblob.sentiment
```

```python
# could we run this and then graph the results 
#(i.e. showing change in sentiment over the duration of the novel)??
for sentence in OTblob.sentences:
    print(sentence)
    print(sentence.sentiment)
```

## Word & Noun Phrase Frequencies

```python
# find single word count, case insensitive 
OTblob.word_counts['oliver']
```

```python
# http://textblob.readthedocs.io/en/dev/quickstart.html#get-word-and-noun-phrase-frequencies

counts = OTblob.word_counts
type(counts)
print(counts)
```

```python
# Find top word frequencies
c = Counter(counts)
c.most_common(40)
```

```python
#OTblob.noun_phrases.count()
noun_phrase = OTblob.np_counts
print(noun_phrase)
```

```python
np = Counter(noun_phrase)
print(np.most_common(20))
```
